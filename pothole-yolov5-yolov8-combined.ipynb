```python
## Prepare lists of Train, Test and Valid image files

def is_Train(filePath):
    if (filePath.find('/train/images/') > 0 ) :
        return True
    else :
        return False

def is_Test(filePath):
    if (filePath.find('/test/images/') > 0 ) :
        return True
    else :
        return False
    
def is_Valid(filePath):
    if (filePath.find('/valid/images/') > 0 ) :
        return True
    else :
        return False
    
trainFileList = list(filter(lambda x: is_Train(x), image_list))
testFileList = list(filter(lambda x: is_Test(x), image_list))
validFileList = list(filter(lambda x: is_Valid(x), image_list))

# trainFileList[: 5]
testFileList[: 5]
# validFileList[: 5]
```
